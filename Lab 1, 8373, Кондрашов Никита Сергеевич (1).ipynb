# Шахматные онлайн-партии

Предметная область - Спорт
Источник - https://www.kaggle.com/datasets/ulrikthygepedersen/online-chess-games
Характер данных - Реальные

В датасете представлены:
game_id - Идентификатор игры (int),
rated - Игра влияет на рейтинг (int, 0 - Нет, 1 - Да),
turns - Количество ходов (int),
victory_status - Условие победы (string),
winner - Цвет победителя (int, Черные - 0, Белые - 1, Ничья - 0.5),
time_increment - Режим часов (string),
white_id - Идентификатор игрока за белые фигуры (string),
white_rating - Рейтинг игрока за белые фигуры (int),
black_id - Идентификатор игрока за черные фигуры (string),
black_rating - Рейтинг игрока за черные фигуры (int).


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import csv
```

```python
games = pd.read_csv('chess_games.csv')
games.head()
```

## Game_id (Идентификатор игры)

Среднее значение:

```python
averageGame_id = games['game_id'].mean()
print(averageGame_id)
```

СКО:

```python
sdGame_id = games['game_id'].std()
print(sdGame_id)
```

```python
plt.figure(figsize=(10, 5)) 
plt.hist (games['game_id'], bins=25, edgecolor='black');
plt.title('game_id distribution');
plt.xlabel("game_id");
plt.ylabel("Quantity");
```

```python
ax = sns.boxplot(x=games.game_id)
```
